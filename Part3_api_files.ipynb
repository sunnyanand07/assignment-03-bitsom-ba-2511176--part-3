```python
# Task 2: API Integration

import requests

#Step 1
url = "https://dummyjson.com/products?limit=20"
response = requests.get(url)
products = response.json()["products"]

print("ID | Title                     | Category      | Price   | Rating")
print("-"*70)

for p in products:
    print(f"{p['id']} | {p['title'][:25]:25} | {p['category']:12} | ${p['rating']}")

# Step 2
filtered = [p for p in products if p["rating"] >= 4.5]
filtered.sort(key=lambda x: x["price"], reverse=True)

print("\nFiltered products:\n")
for p in filtered:
    print(f"{p['title']} - ${p['price']} - Rating: {p['rating']}")

# Step 3
url2 = "https://dummyjson.com/products/category/laptops"
laptops = requests.get(url2).json()["products"]

print("/nLaptop Products:/n")
for p in laptops:
    print(f"{p['title']} - ${p['price']}")

# Step 4
url3 = "https://dummyjson.com/products/add"

new_products = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

response3 = requests.post(url3, json=new_products)

print("\nPOST Response:\n")
print(response3.json())
```

```python
# Task 3: Exception Handling


import requests

# ==============================
# Part A - Guarded Calculated
# ==============================

def safe_divide(a, b ):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

print("Part A Output:")
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))

# ==============================
# Part B - Guarded File Reader
# ==============================

def read_file_safe(filename):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            content = f.read()
            return content
    except FileNotFoundError:
        print(f"Error: File '{filename}' notn found.")
    finally:
        print("File operation attempt complete")

print("\nPart B Output:")
print(read_file_safe("python_notes.txt"))
print(read_file_safe("ghost_file.txt"))

# ==============================
# Part C - Robust API Call

print("\nPart C output:")

# GET request
try:
    url = "https://dummyjson.com/products?limit=5"
    response = requests.get(url, timeout=5)
    data = response.json()
    print("GET Success:", data["products"][0]["title"])
except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print("Error:", e)

# POST request
try:
    url_post = "https://dummyjson.com/products/add"
    new_product = {
        "title": "My Custom Product",
        "price": 999,
        "category": "electronics",
        "description": "A product I created via API"
    }

    response = requests.post(url_post, json=new_product, timeout=5)
    print("POST Success:", response.json())

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print("Error:", e)

# =================================
# Part D - Input Validation Loop
# =================================

print("\nPart D Output:")

while True:
    user_input = input("Enter product ID (1-100) or 'quit': ")

    if user_input.lower() == "quit":
        print("Exiting program...")
        break

    try:
        product_id = int(user_input)

        if product_id < 1 or product_id > 100:
            print("Invalid ID. Enter between 1-100.")
            continue

        url = f"https://dummyjson.com/products/{product_id}"
        response = requests.get(url, timeout=5)

        if response.status_code == 404:
            print("Product not found.")
        elif response.status_code == 200:
            data = response.json()
            print(f"{data['title']} - ${data['price']}")

    except ValueError:
        print("Invalid input. Please enter a number.")

    except requests.exceptions.ConnectionError:
        print("Connection failed. Check internet.")

    except requests.exceptions.Timeout:
        print("Request timed out.")

    except Exception as e:
        print("Error:", e)

    except ValueError:
        print("Invalid input. Please enter a number.")

```

```python
# Task 4: Logging

import requests
from datetime import datetime

# =====================================
# Logger Function
# =====================================
def log_error(function_name, error_type, message):
    """
    Writes a formatted error log entry to error_log.txt
    """
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    log_entry = f"[{timestamp}] ERROR in {function_name}: {error_type} — {message}\n"

    # Append mode so logs are not overwritten
    with open("error_log.txt", "a", encoding="utf-8") as file:
        file.write(log_entry)


# =====================================
# 1. Trigger ConnectionError
# =====================================
def fetch_products():
    try:
        url = "https://this-host-does-not-exist-xyz.com/api"
        response = requests.get(url, timeout=5)
    except requests.exceptions.ConnectionError:
        print("ConnectionError triggered")
        log_error(
            "fetch_products",
            "ConnectionError",
            "No connection could be made"
        )


# =====================================
# 2. Trigger HTTP 404 (Manual Handling)
# =====================================
def lookup_product():
    try:
        url = "https://dummyjson.com/products/999"  # Invalid ID
        response = requests.get(url, timeout=5)

        # HTTP errors are NOT exceptions → check status_code
        if response.status_code != 200:
            print("HTTP error triggered")
            log_error(
                "lookup_product",
                "HTTPError",
                f"{response.status_code} Not Found for product ID 999"
            )

    except requests.exceptions.ConnectionError:
        log_error("lookup_product", "ConnectionError", "No connection")
    except requests.exceptions.Timeout:
        log_error("lookup_product", "Timeout", "Request timed out")
    except Exception as e:
        log_error("lookup_product", "UnknownError", str(e))


# =====================================
# Execute Functions
# =====================================
fetch_products()
lookup_product()


# =====================================
# 3. Read and Display Log File
# =====================================
print("\n===== Error Log Content =====\n")

try:
    with open("error_log.txt", "r", encoding="utf-8") as file:
        print(file.read())
except FileNotFoundError:
    print("Error: Log file not found.")
```
